# 8회 기출

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/inf/refs/heads/main/p4/8_3/churn.csv")
```

```python
print(df.info())
y = df['Churn']
X = df.drop('Churn',axis=1)

import statsmodels.api as sm
X = sm.add_constant(X)

model = sm.Logit(y, X)
result = model.fit()
print(len(result.pvalues[result.pvalues >= 0.05])) # 4
remove = result.pvalues[result.pvalues >=0.05].index.tolist()
print(remove)
X = X.drop(remove,axis=1)
print(X.info())

remodel = sm.Logit(y, X)
result2 = remodel.fit()
print(round(result2.params[result2.pvalues<0.05].sum(),3))

import numpy as np
print(round(np.exp(result2.params['DataUsage']*5),3))
```

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/inf/refs/heads/main/p4/8_3/piq.csv")
```

```python
print(df.info())
y = df['PIQ']
X = df.drop('PIQ', axis=1)

import statsmodels.api as sm
X = sm.add_constant(X)
model = sm.OLS(y, X)
result = model.fit()
print(result.summary())
print(round(result.params[1],3))
print(round(result.rsquared,2))

new_X = pd.DataFrame(
    [[1,90,70,150]], columns=X.columns
)
print(new_X)
y_pred = result.predict(new_X)
print(int(y_pred))
```

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/inf/refs/heads/main/p4/7_3/clam.csv")
```

```python
print(df)

train = df.iloc[:210]
test = df.iloc[210:]
print(train.shape, test.shape)

X_train1 = train['weight']
y_train1 = train['gender']
X_test1 = test['weight']
y_test1 = test['gender']

import statsmodels.api as sm
import statsmodels.api as sm
X_train1 = sm.add_constant(X_train1)
X_test1 = sm.add_constant(X_test1)

model1 = sm.Logit(y_train1, X_train1)
result1 = model1.fit()
print(result1.summary())

import numpy as np
print(round(np.exp(result1.params['weight']),4))

X_train = train.drop(['gender'],axis=1)
y_train = train['gender']
X_test = test.drop(['gender'],axis=1)
y_test = test['gender']

import statsmodels.api as sm
import statsmodels.api as sm
X_train = sm.add_constant(X_train)
X_test = sm.add_constant(X_test)

model2 = sm.Logit(y_train, X_train)
result2 = model2.fit()
print(result2.summary())

print(result2.llf)
print(round(result2.llf * (-2),2))

result3 = result1.predict(X_test1)
import numpy as np
result3 = np.where(result3>=0.5,1,0)
print(result3)
from sklearn.metrics import accuracy_score
score = accuracy_score(y_test1,result3)
print(round(1-score,3))

```

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/inf/refs/heads/main/p4/7_3/system.csv")
```

```python
print(df.corr().iloc[1:,0].abs().idxmax())
print(round(df.corr().iloc[1,0],4))
print(df.describe())

y = df['ERP']
X = df.drop('ERP', axis=1)

import statsmodels.api as sm
X = sm.add_constant(X)

model = sm.OLS(y, X)
result = model.fit()
print(result.summary())

print(round(result.rsquared,3))
print(round(result.pvalues.max(),3))
```
