```python
import pandas as pd
train = pd.read_csv("https://raw.githubusercontent.com/lovedlim/inf/refs/heads/main/p4/6_2/energy_train.csv")
test = pd.read_csv("https://raw.githubusercontent.com/lovedlim/inf/refs/heads/main/p4/6_2/energy_test.csv")

```

```python
print(train.shape, test.shape)
print(train.isna().sum(), test.isna().sum())

print(train.info())
print(train.head())
print(train.nunique())

y = train['Heat_Load']
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
y_encoded = encoder.fit_transform(y)

X = pd.concat([train, test], axis=0)
X = X.drop('Heat_Load', axis=1)
print(X)
object_cols = X.select_dtypes(include="object").columns
print(object_cols)
for col in object_cols:
  encoder_x = LabelEncoder()
  X[col] = encoder_x.fit_transform(X[col])
print(X)

n_train = len(train)
train_df = X.iloc[:n_train]
test_df = X.iloc[n_train:]

from sklearn.model_selection import train_test_split
X_train,X_test, y_train,  y_test = train_test_split(train_df, y_encoded, test_size=0.2, random_state=42)

print(y.value_counts()/len(y))

from sklearn.ensemble import RandomForestClassifier
model =RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)
pred_y_rf = model.predict(X_test)

from sklearn.metrics import f1_score
score = f1_score(y_test, pred_y_rf, average="macro")

print(score)

y_pred = model.predict(test_df)
y_pred_decoder = encoder.inverse_transform(y_pred)
print(y_pred_decoder.shape)
print(test.shape)
print(y_pred_decoder)

result = pd.DataFrame({
    "y_pred": y_pred_decoder
})
print(result)

result.to_csv('result.csv', index=False)
```
